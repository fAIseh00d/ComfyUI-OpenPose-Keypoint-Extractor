```python
import matplotlib.pyplot as plt

set_range=128
DIV=64

x = list(range(0, set_range, 1))
#min_xy = list(map(lambda x: max(round(x/DIV-0.5)*DIV, 0), x))
##wid_he = list(map(lambda x: max(round(x/DIV+0.5)*DIV, 64), x))

min_xy = list(map(lambda x: max(((x-1)//DIV)*DIV, 0), x))
wid_he = list(map(lambda x: max((x//DIV+1)*DIV, 64), x))

plt.plot(x, min_xy)
plt.plot(x, wid_he)
plt.plot(x, x)

plt.ylim(ymin=0, ymax=set_range+8)
plt.xlim(xmin=0, xmax=set_range+8)
plt.show()
```
